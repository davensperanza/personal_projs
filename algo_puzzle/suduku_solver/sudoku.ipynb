```python
# definiamo griglia iniziale
N = 9 #NxN
grid = [[0 for _ in range(N)] for _ in range(N)]
grid[1][1]
```

```python
# posto dove piazzare numero
def first_0(grid):
    for r in range(N):
        for c in range(N):
            if grid[r][c] == 0:
                return r,c
    return None


def findTry(grid):
    mappa = [[-1 for _ in range(N)] for _ in range(N)]
    for r in range(N):
        for c in range(N):
            if grid[r][c] == 0:
                count = 0
                for n in range(1,N+1):
                    if val_values(grid,r,c,n):
                        count += 1
                mappa[r][c] = count
    return mappa


def findLeast(mappa):
    actual_best = N
    count = 0
    a = 0
    b = 0
    for r in range(N):
        for c in range(N):
            if mappa[r][c] == -1:
                continue
            if mappa[r][c] == 0:
                return False
            if mappa[r][c] <= actual_best:
                actual_best = mappa[r][c]
                # if actual_best == 1:
                #     res = (r,c)
                #     return res
                a = r
                b = c
                if actual_best == 1:
                    count += 1
    return a, b, count
            
# check numeri feasible
def val_values(grid, r, c, test): # gli passo riga e colonna dello 0 trovato
    # riga
    check_r = 0
    if test not in grid[r]:
        check_r = 1
    else:
        return False
    
    # colonna
    check_c = 1
    for i in range(N):
        if test == grid[i][c]:
            check_c = 0
    if check_c == 0:
        return False
        
    check_b = 1
    # 3x3
    b = localization(r,c)
    for i in range(b[0],b[0]+3):
        for j in range(b[1],b[1]+3):
            if test == grid[i][j]:
                check_b = 0

    if check_b+check_c+check_r == 3:
        return True
    else:
        return False
    

    
def localization(r,c):
    primo = [0,1,2]
    sec = [3,4,5]
    v = [r,c]
    b = []
    for el in v:
        if el in primo:
            b.append(0)
        elif el in sec:
            b.append(3)
        else:
            b.append(6)

    return b


```

```python
# risolviamo
# annido soluzione così so il punto sbagliato quando non torna più

def solveFast(grid, count=0):
    if count < 2:
        mappa = findTry(grid)
    else: 
        count -= 1
    least = findLeast(mappa)
    if least is False:
        return False
    if least is None:
        return True # risolto
    else:
        r, c, count = least

    for num in range(1,N+1):
        if val_values(grid,r,c,num):
            grid[r][c] = num
            if solveFast(grid):
                return True
            grid[r][c] = 0 # il valore era sbagliato e lo riaggiorno a 0, da fare dopo

    return False

def solveSlow(grid):
    zero = first_0(grid)
    if zero == None:
        return True # risolto
    else:
        r,c = zero

    for num in range(1,N+1):
        if val_values(grid,r,c,num):
            grid[r][c] = num
            if solveSlow(grid):
                return True
            grid[r][c] = 0 # il valore era sbagliato e lo riaggiorno a 0, da fare dopo

    return False
```
