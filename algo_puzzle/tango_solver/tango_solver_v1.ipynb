```python
import copy
```

```python
d = 2      # dimensione griglia pari
grid = [[None for _ in range(d)] for _ in range(d)]
grid
```

```python
def rows(g):
    return g

def cols(g):
    return [list(riga) for riga in zip(*g)]

def check_tr(g):
    # Controllo righe
    for i in range(d):
        row = rows(g)[i]
        for k in range(d-2):
            if None not in row[k:k+3]:
                if sum(row[k:k+3]) >=3 or sum(row[k:k+3]) == 0:
                    return False
    
    # Controllo colonne
    for j in range(d):
        col = cols(g)[j]
        for k in range(d-2):
            if None not in col[k:k+3]:
                if sum(col[k:k+3]) >=3 or sum(col[k:k+3]) == 0:
                    return False
    return True

def check_equal_01(g):
    # Controllo righe
    for i in range(d):
        row = rows(g)[i]
        if (None not in row) and sum(row) != d/2:
            return False
    # Controllo colonne
    for j in range(d):
        column = cols(g)[j]
        if (None not in column) and sum(column) != d/2:
            return False
    return True

def is_valid_grid(g):
    return check_equal_01(g) and check_tr(g)

def insert_grid():
    for i in range(d):
        inp = input(f"Riga {i}-esima")
        for j in range(d):
            if inp[j] != 'n': 
                grid[i][j] = int(inp[j])
```

```python
insert_grid()

for i in range(d):
    print(cols(grid)[i])
```

```python
def SolveTango(g):    
    prev_count_none = -1
    while any(None in row for row in g) and is_valid_grid(g):
        count_none = sum(cell is None for row in grid for cell in row)
        if prev_count_none == count_none:
            print("Grid is not solvable")
            break
        else:
            prev_count_none = count_none
        for i in range(d):
            for j in range(d):
                if g[i][j] == None:
                    future_grid = copy.deepcopy(g)
                    future_grid[i][j] = 1
                    if (not check_equal_01(future_grid)) or (not check_tr(future_grid)):
                        g[i][j] = 0
                    else:
                        future_grid = copy.deepcopy(g)
                        future_grid[i][j] = 0
                        if (not check_equal_01(future_grid)) or (not check_tr(future_grid)):
                            g[i][j] = 1
                        else:
                            pass
                else: pass
```

```python
if is_valid_grid(grid):
    SolveTango(grid)            # Solve grid
else:
    print("Cannot solve invalid grid")
for i in range(d):
    print(grid[i])
```
